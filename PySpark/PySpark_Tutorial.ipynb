### PySpark handling missing values

```python
# drop the rows with null values
df_spark.na.drop().show()
```

```python
 # drop thre rows using  how, thresh, subset
df_spark.na.drop(how = "any").show() # drop the rows if any of the value is null
```

```python
df_spark.na.drop(how = "all").show() # drop the rows if all of the value is null
```

```python
df_spark.na.drop(how = "all", thresh= 2).show() # drop the rows if there are less than 2 non-null values
```

```python
df_spark.na.drop(subset = ["Experience"]).show() # drop the rows if there are null values in the "Experience" column 
```

```python
# replace the null values all the columns
df_spark.na.fill("N\A").show() # replace the null values with "Unknown", only for the string columns
# df_spark.na.fill(0).show() # replace the null values with 0, only for the numeric columns
```

```python
# load imputer 
from pyspark.ml.feature import Imputer
# create an imputer object for the numeric columns, with the strategy of mean
imputer = Imputer(inputCols=["Age", "Salary", "Experience"], outputCols=["Age_imputed","Salary_imputed", "Experience_imputed"], strategy="mean") # if want to fill with median, then use strategy = "median" 
# fit the imputer object to the dataframe
imputer.fit(df_spark).transform(df_spark).show()
```

### Simple linear regression using PySpark

```python
# loading test1.csv
training = spark.read.csv("/Users/sahadathossain/Desktop/Grow_Skills/Data_Engineering_Learning/PySpark/test1.csv", header= True, inferSchema= True)
training.show()
```

```python
training.printSchema()
```

```python
training.columns
```

[Age, Experice] ---> new future ---> independent features

```python
# load vector assembler
from pyspark.ml.feature import VectorAssembler
# create a vector assembler object
assembler = VectorAssembler(inputCols=["age", "Experience"], outputCol="independent features")
output = assembler.transform(training)
output.show()   
```

```python
# keep only Salary and independent features columns
finalized_data = output.select("Salary", "independent features")
```

```python
# load linear regression model
from pyspark.ml.regression import LinearRegression
# train and test data split
train_data, test_data = finalized_data.randomSplit([0.75, 0.25])      
# create a linear regression object
lr = LinearRegression(featuresCol= "independent features", labelCol= "Salary")
# fit the model to the training data
model = lr.fit(train_data)
```

```python
# coefficients and intercept of the model
print("Coefficients: ", model.coefficients)
print("Intercept: ", model.intercept)
```

```python
# predictions   
pred_result = model.evaluate(test_data)
pred_result.predictions.show()
```

```python
# mse, rmse, r2, mae
print("MSE: ", pred_result.meanSquaredError)
print("RMSE: ", pred_result.rootMeanSquaredError)
print("R2: ", pred_result.r2)
print("MAE: ", pred_result.meanAbsoluteError)
```
